```python
class Solution(object):
    def findMin(self, nums):
        low=0
        l=len(nums)
        high=l-1
        if l==1:
            return(nums[0])
        while(low<=high):
            mid=(low+high)//2
            if mid==0:
                prevmid=l-1
            else:
                prevmid=mid-1
            if nums[mid]<nums[prevmid]:
                return nums[mid]
            elif nums[mid]>nums[l-1]:
                low=mid+1
            else:
                high=mid-1
```

```python
class Solution(object):
    def setZeroes(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: None Do not return anything, modify matrix in-place instead.
        """
        col0=1
        for i in range(len(matrix)):
            for j in range(len(matrix[0])):
                if matrix[i][j]==0:
                    matrix[i][0]=0
                    if(j!=0):
                        matrix[0][j]=0
                    else:
                        col0=0
        for i in range(1,len(matrix)):
            for j in range(1,len(matrix[0])):
                if(matrix[i][0]==0 or matrix[0][j]==0):
                    matrix[i][j]=0
        if matrix[0][0]==0:
            for i in range(len(matrix[0])):
                matrix[0][i]=0
        if col0==0:
            for j in range(len(matrix)):
                matrix[j][0]=0
```

```python
class Solution(object):
    def generate(self, numRows):
        """
        :type numRows: int
        :rtype: List[List[int]]
        """
        L=[]
        for i in range(1,numRows+1):
            l=[1]
            a=1
            for j in range(1,i):
                a*=i-j
                a//=j
                #print(a)
                l.append(a)
            L.append(l)
        return L
```

```python
class Solution(object):
    def maxSubArray(self, nums):
        max1=float('-inf')
        sum=0
        for i in range(len(nums)):
            sum+=nums[i]
            #print(max1,sum)
            if sum>max1:
                max1=sum
            if sum<0:
                sum=0
        return max1
```

```python
class Solution(object):
    def maxProduct(self, nums):
        max1=float('-inf')
        prod1=1
        prod2=1
        for i in range(len(nums)):
            j=len(nums)-i-1
            prod1*=nums[i]
            max1=max(prod1,max1)
            if prod1==0:
                prod1=1
            prod2*=nums[j]
            max1=max(prod2,max1)
            if prod2==0:
                prod2=1
        return max1
```

```python
class Solution(object):
    def rotate(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: None Do not return anything, modify matrix in-place instead.
        """
        for i in range(len(matrix)):
            for j in range(i):
                matrix[i][j],matrix[j][i]=matrix[j][i],matrix[i][j]
        for i in range(len(matrix)):
            matrix[i]=matrix[i][::-1]
```

```python
class Solution(object):
    def merge(self, nums1, m, nums2, n):
        if m==0:
            for i in range(len(nums1)):
                nums1.pop()
            for j in range(n):
                nums1.append(nums2[j])
            #print(nums1)
            return nums1
        if n==0:
            return nums1
        left=m-1
        right=0
        while(left>=0 and right<n):
            if(nums1[left]>nums2[right]):
                nums1[left],nums2[right]=nums2[right],nums1[left]
                left-=1
                right+=1
            else:
                break
        k=0
        while(m<len(nums1)):
            nums1[m]=nums2[k]
            m+=1
            k+=1
        nums1.sort()
        return nums1
```

```python



from typing import List

def findMissingRepeatingNumbers(a: [int]) -> [int]:
    n = len(a)  # size of the array

    # Find Sn and S2n:
    SN = (n * (n + 1)) // 2
    S2N = (n * (n + 1) * (2 * n + 1)) // 6

    # Calculate S and S2:
    S, S2 = 0, 0
    for i in range(n):
        S += a[i]
        S2 += a[i] * a[i]

    # S-Sn = X-Y:
    val1 = S - SN

    # S2-S2n = X^2-Y^2:
    val2 = S2 - S2N

    # Find X+Y = (X^2-Y^2)/(X-Y):
    val2 = val2 // val1

    # Find X and Y: X = ((X+Y)+(X-Y))/2 and Y = X-(X-Y),
    # Here, X-Y = val1 and X+Y = val2:
    x = (val1 + val2) // 2
    y = x - val1

    return [x, y]

if __name__ == '__main__':
    a = [3, 1, 2, 5, 4, 6, 7, 5]
    ans = findMissingRepeatingNumbers(a)
    print("The repeating and missing numbers are: {", ans[0], ", ", ans[1], "}\n")

```

```python

from typing import List
import math

def merge(arr : List[int], low : int, mid : int, high : int) -> int:
    temp = []   # temporary array
    left = low  # starting index of left half of arr
    right = mid + 1 # starting index of right half of arr

    cnt = 0     # Modification 1: cnt variable to count the pairs

    # storing elements in the temporary array in a sorted manner
    while (left <= mid and right <= high):
        if (arr[left] <= arr[right]):
            temp.append(arr[left])
            left += 1
        else:
            temp.append(arr[right])
            cnt += (mid - left + 1)  # Modification 2
            right += 1

    # if elements on the left half are still left
    while (left <= mid):
        temp.append(arr[left])
        left += 1

    # if elements on the right half are still left
    while (right <= high):
        temp.append(arr[right])
        right += 1

    # transfering all elements from temporary to arr
    for i in range(low, high + 1):
        arr[i] = temp[i - low]

    return cnt   # Modification 3

def mergeSort(arr : List[int], low : int, high : int) -> int:
    cnt = 0
    if low >= high:
        return cnt
    mid = math.floor((low + high) / 2)
    cnt += mergeSort(arr, low, mid)    # left half
    cnt += mergeSort(arr, mid + 1, high)  # right half
    cnt += merge(arr, low, mid, high)  # merging sorted halves
    return cnt

def numberOfInversions(a : List[int], n : int) -> int:
    # Count the number of pairs:
    n = len(a)
    # Count the number of pairs:
    return mergeSort(a, 0, n - 1)

if __name__ == "__main__":
    a = [5, 4, 3, 2, 1]
    n = 5
    cnt = numberOfInversions(a, n)
    print("The number of inversions are:", cnt)

```
