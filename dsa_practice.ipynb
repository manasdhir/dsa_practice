```python
class Solution(object):
    def findMin(self, nums):
        low=0
        l=len(nums)
        high=l-1
        if l==1:
            return(nums[0])
        while(low<=high):
            mid=(low+high)//2
            if mid==0:
                prevmid=l-1
            else:
                prevmid=mid-1
            if nums[mid]<nums[prevmid]:
                return nums[mid]
            elif nums[mid]>nums[l-1]:
                low=mid+1
            else:
                high=mid-1
```

```python
class Solution(object):
    def setZeroes(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: None Do not return anything, modify matrix in-place instead.
        """
        col0=1
        for i in range(len(matrix)):
            for j in range(len(matrix[0])):
                if matrix[i][j]==0:
                    matrix[i][0]=0
                    if(j!=0):
                        matrix[0][j]=0
                    else:
                        col0=0
        for i in range(1,len(matrix)):
            for j in range(1,len(matrix[0])):
                if(matrix[i][0]==0 or matrix[0][j]==0):
                    matrix[i][j]=0
        if matrix[0][0]==0:
            for i in range(len(matrix[0])):
                matrix[0][i]=0
        if col0==0:
            for j in range(len(matrix)):
                matrix[j][0]=0
```

```python
class Solution(object):
    def generate(self, numRows):
        """
        :type numRows: int
        :rtype: List[List[int]]
        """
        L=[]
        for i in range(1,numRows+1):
            l=[1]
            a=1
            for j in range(1,i):
                a*=i-j
                a//=j
                #print(a)
                l.append(a)
            L.append(l)
        return L
```

```python
class Solution(object):
    def nextPermutation(self, nums):
        ind=-1
        for i in range(len(nums)-2,-1,-1):
            if  nums[i]<nums[i+1]:
                ind=i
                break
        if ind==-1:
            #if max(nums)==nums[0]:
            nums.reverse()
            return nums
        for i in range(len(nums)-1,ind,-1):
            if(nums[i]>nums[ind]):
                nums[i],nums[ind]=nums[ind],nums[i]
                break
        nums[ind+1:]=nums[ind+1:][::-1]
        return nums
```

```python
class Solution(object):
    def maxSubArray(self, nums):
        max1=float('-inf')
        sum=0
        for i in range(len(nums)):
            sum+=nums[i]
            #print(max1,sum)
            if sum>max1:
                max1=sum
            if sum<0:
                sum=0
        return max1
```

```python
class Solution(object):
    def maxProduct(self, nums):
        max1=float('-inf')
        prod1=1
        prod2=1
        for i in range(len(nums)):
            j=len(nums)-i-1
            prod1*=nums[i]
            max1=max(prod1,max1)
            if prod1==0:
                prod1=1
            prod2*=nums[j]
            max1=max(prod2,max1)
            if prod2==0:
                prod2=1
        return max1
```

```python
#O(n)+O(n) approach;double pass
class Solution(object):
    def sortColors(self, nums):
        d={0:0,1:0,2:0}
        for i in range(len(nums)):
            d[nums[i]]+=1
        for j in range(len(nums)):
            if d[0]>0:
                nums[j]=0
                d[0]-=1
                pass
            elif d[1]>0:
                nums[j]=1
                d[1]-=1
                pass
            elif d[2]>0:
                nums[j]=2
                d[2]-=1
                pass
```

```python
#O(n) approach;single pass
class Solution(object):
    def sortColors(self, nums):
        low=0
        high=len(nums)-1
        mid=0
        while(mid<=high):
            if(nums[mid]==0):
                nums[mid],nums[low]=nums[low],nums[mid]
                mid+=1
                low+=1
            elif(nums[mid]==1):
                mid+=1
            else:
                nums[mid],nums[high]=nums[high],nums[mid]
                high-=1
        return nums
```

```python
class Solution(object):
    def maxProfit(self, prices):
        min1=float("inf")
        pro=0
        for i in range(len(prices)):
            min1=min(min1,prices[i])
            pro=max(prices[i]-min1,pro)
        return pro
```

```python
class Solution(object):
    def maxProfit(self, prices):
        profit=0
        for i in range(1,len(prices)):
            if(prices[i]>prices[i-1]):
                profit+=prices[i]-prices[i-1]
        return profit
```
